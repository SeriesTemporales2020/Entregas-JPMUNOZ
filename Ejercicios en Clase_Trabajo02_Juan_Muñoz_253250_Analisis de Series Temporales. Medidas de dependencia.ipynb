## Medidas de dependencia. Covarianza, autocorrelación, correlación cruzada.



## Ejercicios

1. Simular una señal de ruido blanco y estimar su autocorrelación. Probar con $n=50$ y $n=200$ datos y discutir las diferencias.
2. Simular el proceso de media móvil visto anteriormente y estimar su autocorrelación. Probar nuevamente con $n=50$ y $n=200$ datos y discutir las diferencias.
3. Simular el proceso autorregresivo visto anteriormente y repetir la pregunta anterior. Discutir las diferencias entre los 3 procesos.

Sugerencia: usar los comandos `filter` y `acf` de R.

### Ejercicio 1

```r
install.packages("devtools")
devtools::install_github("nickpoison/astsa")
#install.packages("astsa")
#library(astsa)
```

```r
x_ta = rnorm(n = 50, mean = 0 , sd = 1) #Ruido blanco para n=50
x_tb = rnorm(n = 200, mean = 0 , sd = 1) #Ruido blanco para n=200
```

```r
tsplot(x_ta, main = "xta", col = 6, xlab = " Time" , ylab = "Value")
tsplot(x_tb, main = "xtb", col = 3, xlab = " Time" , ylab = "Value")
```

```r
acf(x_ta, main = "Autocorrelación de Ruido blanco n = 50") #Autocorrelación de Ruido blanco para n= 50
acf(x_tb, main = "Autocorrelación de Ruido blanco n = 200") #Autocorrelación de Ruido blanco para n= 200
```

$$Observación:$$ A mayor número de muestras, más visible se hace determinar el Ruido Blanco con la función ACF.los valores de cada Lag, claramente se encuentra debajo e la inea azul.

### Ejercicio 2

```r
v_ta = filter(x_ta, sides=2, rep(1/3,3))[!is.na(filter(x_ta, sides = 2, filter = rep(1/3,3)))]
v_tb = filter(x_tb, sides=2, rep(1/3,3))[!is.na(filter(x_tb, sides = 2, filter = rep(1/3,3)))]
tsplot(v_ta, main = "Media móvil n = 50", col = 6)
tsplot(v_tb, main = "Media móvil n = 200", col = 3)
```

```r
acf(v_ta, main = "Autocorrelación de Media móvil n = 50")
acf(v_tb, main = "Autocorrelación de Media móvil n = 200")
```

```r
acf(v_ta, main = "Autocorrelación de Media móvil n = 50", plot = FALSE)
acf(v_tb, main = "Autocorrelación de Media móvil n = 200", plot = FALSE)
```

$$Observación:$$ En las dos gráficas se observa que el descenso de los valores en las gráfca del ACF  es caracterisitico de la Media móvil. Tambien se puede obervar que, al igual que en las grafiacas de ACF del ruido blanco, a mayor numero de Lags se observa con mayor facilidad su autocorrelación."}]

### Ejercicio 3

```r
ax_ta = filter(x_ta, filter=c(1,-.9), method="recursive")
tsplot(ax_ta, main="autoregression de Ruido blanco n=50", col = 6)
```

```r
ax_tb = filter(x_tb, filter=c(1,-.9), method="recursive")
tsplot(ax_tb, main="autoregression de Ruido blanco n=200", col = 3)
```

```r
par(mfrow=c(3,2))
acf(x_ta, main = "Autocorrelación de Ruido blanco n = 50") #Autocorrelación de Ruido blanco para n= 50
acf(x_tb, main = "Autocorrelación de Ruido blanco n = 200") #Autocorrelación de Ruido blanco para n= 200
acf(v_ta, main = "Autocorrelación de Media móvil n = 50") #Autocorrelación de Media móvil n= 50
acf(v_tb, main = "Autocorrelación de Media móvil n = 200") #Autocorrelación de Media móvil n= 200
acf(ax_ta, main = "Autocorrelación de proceso autorregresivo n = 50") #Autocorrelación de Autorregresivo n=50
acf(ax_tb, main = "Autocorrelación de proceso autorregresivo n = 200") #Autocorrelación de Autorregresivo n=200
```

$$Observación:$$ En los 3 ejercicios se ha realizado el mismo procedimiento de analizar la Autocorrelación para cada uno, un tipo de señal diferente con un número diferente de Lags, Señal de Ruido Gaussiano, Señal de Media móvil y Señal de un proceso Autorregresivo; para todas estas usando la función "ACF". Con estas 3 comparaciones, se identifica claramente que a mayor número de lags, es más visible esta caracterisitica. 
Luego, a partir de esto, se puede llegar a identificar con esta función estos tipos de señales.
Sabiendo que:

La señal de Ruido blanco, tiene el valor de Autoccorrelación 1 en el lag 0, y para los demás lags, valores no siginificativos que se encuentran debajo de la frnaja entre las lineas azúles.

La señal de Medía móvil, tiene la caracteristica que tiene un decrecimiento a partir del primer lag de forma y apariencia lienal, los demás lags, dentro de las franjas delimitadas de las líneas azúles. 

La señal generada del proceso autorregresivo, muestra que la señal decrece exponencialmente con oscilaciones.
